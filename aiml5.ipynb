```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales.csv")
```

```python
plt.figure()
plt.plot(data['Unit price'],data['Total'],marker='*')
plt.title('Line Plot')
plt.xlabel('Unit price')
plt.ylabel('Total')
plt.show()
```

```python
plt.figure()
plt.bar(data['Quantity'],data['cogs'])
plt.title('Bar Chart')
plt.xlabel('Quantity')
plt.ylabel('cogs')
plt.show()
```

```python
#4 histogran(seaborn)
plt.figure()
plt.hist(data['Quantity'], bins=6)
plt.title('Histogram')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
plt.scatter(data['Quantity'],data['Unit price'])
plt.title('Scatter plot')
plt.xlabel('Quantity')
plt.ylabel('Unit price')
plt.show()
```

```python
#5 Box PLot (seaborn)
plt.figure()
sns.boxplot(x=data['Quantity'], y='Tax 5%' , data=data)
plt.title('Box Plot')
plt.xlabel('Quantity')
plt.ylabel('Tax 5%')
plt.show()
```

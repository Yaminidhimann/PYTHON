```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_salesnew.csv")
```

```python
df=pd.read_csv("supermarket_salesnew.csv")
df.info()
```

```python
df.head()
```

```python
#get the shape of dataframe(rows,columns)
df.shape
```

```python
#get the summary of statistics
df.describe()
```

```python
#Check the missing values
df.isnull()
```

```python
#Total missing values
df.isnull().sum()
```

```python
#Drop rows with missing values(if appropriate)
df.dropna(inplace=True)
```

```python
df.fillna(df['Tax 5%'].mean(),inplace=True)
```

```python
df.head()
```

```python
#Histogram:visualize the distribution of numerical variables
sns.histplot(df["Tax 5%"])
plt.show()
```

```python
sns.boxplot(x='City',y='Tax 5%', data=df)
plt.show()
```

```python
#Scatterplot: Examine relationship between two numercal values
sns.scatterplot(x='cogs',y='Unit price',data=df)
plt.show()
```

```python
#Bar charts : Visualize the distribution of categorial values
sns.countplot(x='Branch',data=df)
plt.show()
```

```python
#correlation Matrix: Understand the relationship between multiple numerical values
af=df[['Unit price','Quantity','Tax 5%','cogs','gross income','Rating']]
corr=af.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#Groupby: Analyze data based on different categories
df.groupby('City')['Quantity'].mean()
```
